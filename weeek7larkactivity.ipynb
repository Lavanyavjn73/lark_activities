# Check whether female and male students scored the same marks.

```python
male_scores=df.loc[df['Gender']=='male']
female_scores=df.loc[df['Gender']=='female']
t,pval=stats.ttest_ind(male_scores['Total Marks'],female_scores['Total Marks'])
print('p value :',pval)
if pval <0.05:
  print(" Null hypothesis is rejected \n","Females and Males score same marks")
else:
  print(" Null hypothesis is accepted \n","Female and Male scores are different")
```

# Whether test preparation helps the students?

```python
tp_none=df[df['Test preparation']=='none']
tp_completed=df[df['Test preparation']=='completed']
tval,pval=stats.ttest_ind(tp_none['Total Marks'],tp_completed['Total Marks'])
print('p value :',pval)
if pval <0.05:
  print(" Null hypothesis is rejected \n","Test preparation helps students to score better")
else:
  print(" Null hypothesis is accepted \n","Test preparation is not useful")
```
